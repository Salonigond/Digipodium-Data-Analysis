```python
import numpy as np
import matplotlib.pyplot as plt
```

get the data/generate data

```python
rainfall = [67,20,10,50,26,59]
months= ['jan','feb','mar','apr','may','jun']
```

```python
plt.style.available
```

```python
plt.style.use('ggplot') #theme
```

visualize the data

```python
plt.bar(months,rainfall,color='#1f77b4')
```

```python
plt.pie(
    rainfall,
    labels=months,
    autopct='%1.1f%%',
)
plt.show()
```

```python
# donut
plt.pie(
    rainfall,
    labels=months,
    autopct='%1.1f%%',
    wedgeprops={'width': 0.4, 'edgecolor': '#6666ff', 'linewidth':2},
    pctdistance=.8
)
plt.show()
```

```python
plt.boxplot(rainfall,
            vert=False,
            showmeans=True,
            meanline=True,
            meanprops={'color': 'red'}
            )
plt.show()

```
